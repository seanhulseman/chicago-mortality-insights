# Clustered Data

```python
clusters = pd.read_csv('../1_data/df_with_topics.csv')
clusters.head()
```

```python
clusters.best_topic_name.value_counts().plot(kind = 'barh', color='royalblue')
plt.xlabel('Number of Occurances')
plt.ylabel('Primary Cause Cluster')
plt.title('Cause of Death Cluster Counts');

plt.savefig('cause_of_death_clusters.png')
```

```python
clusters.inc_day.value_counts().plot(kind = 'barh', color='royalblue')
plt.xlabel('Number of Occurances')
plt.ylabel('Day of Incident')
plt.title('Day of Incident Counts');

plt.savefig('day_of_incidents.png')
```

```python
clusters.death_day.value_counts().plot(kind = 'barh', color='royalblue')
plt.xlabel('Number of Occurances')
plt.ylabel('Day of Death')
plt.title('Day of Death Counts');

plt.savefig('day_of_death.png')
```

```python
clusters.age_range.value_counts().plot(kind = 'barh', color='royalblue')
plt.xlabel('Number of Occurances')
plt.ylabel('Age Rage')
plt.title('Age Range Counts');

plt.savefig('age_range_counts.png')
```

```python
clusters['manne'] = clusters['manner_of_death'].map({
    'ACCIDENT':0,
    'HOMICIDE':1,
    'SUICIDE':2,
})

clusters.head()
```

```python
X = clusters['primary_cause']

tvec = TfidfVectorizer(stop_words = "english", ngram_range = (2,2))

tvec.fit(X)

X = tvec.transform(X)


X_df = pd.DataFrame(X.todense(), columns=tvec.get_feature_names_out())
X_df.sum().sort_values(ascending = False).head(10).plot(kind = 'barh', color = 'lime');

plt.savefig('primary_cause_top_10_n_gram.png')
```

```python
X = clusters['primary_cause']

tvec = TfidfVectorizer(stop_words = "english")

tvec.fit(X)

X = tvec.transform(X)


X_df = pd.DataFrame(X.todense(), columns=tvec.get_feature_names_out())
X_df.sum().sort_values(ascending = False).head(25).plot(kind = 'barh', color = 'lime');

plt.savefig('primary_cause_top_25_single_word.png')
```
