## EDA

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dummy/Medical_Examiner_Case_Archive.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
pd.set_option('display.max_columns', None)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.drop(columns=['case_number','incident_address','location','objectid','chicago_ward'],inplace= True)
```

```python
df['manner_of_death'].value_counts()
```

```python
df.drop(index=df.loc[(df['manner_of_death']== 'NATURAL') |(df['manner_of_death']== 'UNDETERMINED') |(df['manner_of_death']== 'PENDING')].index,inplace=True)
```

```python
df = df[df['manner_of_death'].notna()]
```

```python
606l,60634-3976,LOMINO,606600
```

```python
zip_code = []
for item in df['incident_zip_code']:
    if '-' in str(item):
        zip_code.append(item[:5])
    elif str(item).isnumeric() and len(str(item).strip())!=5:
        zip_code.append(np.nan)
    elif len(str(item).strip()) == 5:
        zip_code.append(str(item))
    else:
        zip_code.append(np.nan)
```

```python
len(zip_code)
```

```python
df['residence_zip'].value_counts().sort_index()
```

```python
zip_code_r = []
for item in df['residence_zip']:
    if '-' in str(item):
        zip_code_r.append(item[:5])
    elif str(item).isnumeric() and len(str(item).strip())!=5:
        zip_code_r.append(np.nan)
    elif len(str(item).strip()) == 5:
        zip_code_r.append(str(item))
    else:
        zip_code_r.append(np.nan)
```

```python
len(zip_code_r)
```

```python
df['incident_zip_code'] = zip_code
```

```python
df['residence_zip'] = zip_code_r
```

```python
df.head()
```

```python
df.to_csv('dataframe.csv')
```
