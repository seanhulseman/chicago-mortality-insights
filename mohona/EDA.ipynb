# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/dataframe.csv')
df.head()
```

```python
df.columns
```

```python
sns.set(style="whitegrid")
# Plot the age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Plot gender distribution
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='gender', palette="Set2")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Plot race distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='race', palette="Set3")
plt.title("Race Distribution")
plt.xlabel("Race")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot manner of death
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='manner_of_death', palette="Set1")
plt.title("Manner of Death Distribution")
plt.xlabel("Manner of Death")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot opioid-related deaths
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='opioid_related', palette="husl")
plt.title("Opioid-Related Deaths")
plt.xlabel("Opioid Related")
plt.ylabel("Count")
plt.show()
```

```python
# Plot age vs. opioid-related deaths
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='opioid_related', y='age', palette="Pastel1")
plt.title("Age vs. Opioid-Related Deaths")
plt.xlabel("Opioid Related")
plt.ylabel("Age")
plt.show()
```

```python
# Calculate and visualize the distribution of age by gender
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='gender', y='age', palette="Set2")
plt.title("Age Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
# Explore the primary causes of death
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='primary_cause', order=df['primary_cause'].value_counts().index[:10], palette="Set3")
plt.title("Top 10 Primary Causes of Death")
plt.xlabel("Count")
plt.ylabel("Primary Cause")
plt.show()
```

```python
# Visualize the distribution of gun-related deaths
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='gun_related', palette="Set1")
plt.title("Gun-Related Deaths")
plt.xlabel("Gun Related")
plt.ylabel("Count")
plt.show()
```

```python
# Plot the relationship between opioid-related deaths and gun-related deaths
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='opioid_related', hue='gun_related', palette="Set2")
plt.title("Opioid-Related vs. Gun-Related Deaths")
plt.xlabel("Opioid Related")
plt.ylabel("Count")
plt.legend(title="Gun Related", labels=["No", "Yes"])
plt.show()
```

```python
# Explore the distribution of ages for COVID-related deaths
plt.figure(figsize=(10, 6))
sns.histplot(data=df[df['covid_related'] == 1], x='age', bins=30, kde=True, color='red', label='COVID-Related')
sns.histplot(data=df[df['covid_related'] == 0], x='age', bins=30, kde=True, color='blue', label='Non COVID-Related')
plt.title("Age Distribution for COVID-Related and Non COVID-Related Deaths")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# Stacked Bar Chart for COVID-Related Deaths by Gender and Race
covid_df = df[df['covid_related'] == 1]  # Filter COVID-related deaths

plt.figure(figsize=(12, 8))
sns.countplot(data=covid_df, x='gender', hue='race', palette="Set2")
plt.title("COVID-Related Deaths by Gender and Race")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Race")
plt.show()
```
